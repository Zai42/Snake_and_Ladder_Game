# Introduction
This AI (Lab) Project is a classic board game (Snake And ladder)

---
**How Game Works:**

In this game there are two players. Each player has to roll a dice on
their turn and they have to move their character on board according to 
the number that comes on dice. Each time new value on dice will be added to previous value to get the new position of player on board.


On specific values there are **Snakes** and **Ladders**.If character's current position is on value of snake he has to move 5 position back 
If character's current position is on value of ladder he has to move 5 position forward.

**First Player to reach 50 will win!!**

---


**Python Concepts used in this project:**



*   Python "Random" Library

    1.   random.randint() function

*   User defined Functions

    1.   posforp1 - Generate Random Dice Value for Player 1
    2.   posforp2 - Generate Random Dice Value for Player 2
    3.   posforplayers - Determine Position of Players on board based on Dice Values
*   If/elif/else
